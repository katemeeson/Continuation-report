```python
print('and rules: ', len(and_rules)) #599 out of 3017 rules are 'and' rules. These will be filtered manually.
print('or rules: ', len(or_rules)) #2418 out of 3017 rules are 'or' rules, and contain those proteins in our overlap so reactions can happen.
```
